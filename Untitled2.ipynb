Part A -- Translation (Conceptual)




For the R code below:

Predict the output\
Translate it into Python\







Compute the result in Python


x <- c(1, 3, 5, 7)

y <- ifelse(x > 4, x^2, x + 1)

mean(y)



Clearly show your Python code and final result.

```python
import numpy as np

# Creating vector
x = np.array([1, 3, 5, 7])

# Applying the vectorized conditional, as we notice in R as per the query above
y = np.where(x > 4, x**2, x + 1)

# Compute mean
result = np.mean(y)

y, result
```

Part B -- Lists vs NumPy Arrays


Create a Python list of numbers 1--5\
Attempt to multiply the list by 2\
Convert the list to a NumPy array and repeat the multiplication





Answer the following questions:

What happened in each case?\
Why does NumPy behave differently?

```python
lst = [1,2,3,4,5]

lst*2
```

```python
import numpy as np

np.array(lst)*2
```

1. lst*2 ---> repeats the same list
2. np.array(lst)*2 ---> seems to properly give us the multiplication result by multiplying every element within the array with 2.

** Python list is using repetition
** Numpy array used vectorized operation like R
