We will estimate the median hour of crime incidents using a random 10% sample and bootstrap simulation (a common nonparametric method).

### Research Question:       



What is the estimated median hour at which crime incidents occur, based on a random sample of the dataset?         

To estimate the median hour of crime incidents, a random sample consisting of approximately 10% of the dataset was drawn. Because we do not assume a specific probability distribution for the variable, a nonparametric approach was used. In particular, a bootstrap simulation was performed by repeatedly resampling from the random sample and calculating the median for each resample. This allows us to approximate the sampling distribution of the median and estimate the central tendency of incident timing.

```python
sample_df = df.sample(frac=0.10, random_state=1)

sample_df.shape
```

```python
sample_median = sample_df["hour"].median()

sample_median
```

```python
boot_medians = []

for i in range(1000):
    boot_sample = sample_df["hour"].sample(frac=1, replace=True)
    boot_medians.append(boot_sample.median())
```

```python
plt.figure(figsize=(8,5))

sns.histplot(boot_medians, bins=20)

plt.xlabel("Bootstrapped Median Hour")
plt.ylabel("Frequency")
plt.title("Bootstrap Distribution of the Median Hour")

plt.show()
```

### Results and Interpretation


Bootstrap simulation produced a distribution of median values which is based on repeated resampling from the 10% of the random sample.
The histogram arising from the data shows the approximate sampling distribution.     


Results show that the median hour at which the incidents occur tends to cluster aroudn a central value indicating a typical time period 
when incidents are most common.
