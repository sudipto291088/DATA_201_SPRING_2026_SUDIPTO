Part A – Core Wrangling (Method Chaining Required)
Using one chained expression, create a summary table that:

Keeps only observations where price > 250000
Keeps only homes with size > 1000
Creates a new variable:
price_per_sqft = price / size
Groups by neighborhood
Computes:
mean of price_per_sqft
median of price_per_sqft
count of homes
Sorts the result by mean price_per_sqft (descending)
Requirements
Do NOT create intermediate variables (df2, df3, etc.)
Use .assign() to create new variables
Use .agg() with named aggregation
Do not use reset_index() (we did not cover it yet)
Output format note
After groupby(), pandas will put neighborhood in the index. That is OK for this assignment. Your final output can look like a table where neighborhood is on the left as the index.

Deliverable
Paste your code
Show the final DataFrame output

```python
(df
 .loc[(df["price"] > 250000) & (df["size"] > 1000)]
 .assign(price_per_sqft=lambda d: d["price"] / d["size"])
 .groupby("neighborhood")
 .agg(
     mean_price_per_sqft=("price_per_sqft", "mean"),
     median_price_per_sqft=("price_per_sqft", "median"),
     count_homes=("price_per_sqft", "size")
 )
 .sort_values("mean_price_per_sqft", ascending=False)
)
```

```python
library(dplyr)

housing %>%
  filter(price > 250000, size > 1000) %>%
  mutate(price_per_sqft = price / size) %>%
  group_by(neighborhood) %>%
  summarise(
    mean_price_per_sqft = mean(price_per_sqft),
    median_price_per_sqft = median(price_per_sqft),
    count_homes = n()
  ) %>%
  arrange(desc(mean_price_per_sqft))
```

Part C – Boolean Logic Debugging
The following code produces an error:

Tasks

Rewrite the filter using .query() instead.

```python
## Fixed Code


df[(df["price"] > 250000) & (df["size"] > 1000)]
```

```python
## Why the error occurs:

## & has higher precedence than >. Hence Py reads it firstly as 250000 & df["size"].
```
